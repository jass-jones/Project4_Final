```python
import pandas as pd
import numpy as np
```

## DATA IMPORT

```python
db = pd.read_csv('Resources/Top10000genre.csv')
top = pd.read_csv('Resources/Top200.csv')
```

## DATA CLEANING

```python
db.head()
```

```python
db.info()
```

```python
db.columns = db.columns.str.replace(' ', '')
db.info()
```

```python
top.head()
```

```python
top.info()
```

```python
top.columns = top.columns.str.lower().str.replace(' ', '_')
top.info()
```

```python
top.rename(columns = {'artist': 'artist_name', "song_name": "track_name", "genre" : "genre_list", "chord": "key", "song_id": "track_id", "duration_(ms)": "duration_ms"}, inplace = True)
top.info()
```

```python
top.drop(columns = ['index'], inplace = True)
top["top_200"] = 1
```

```python
## drop duplicates
top_dedup = top.drop_duplicates(subset = ['track_name', 'artist_name'], keep = 'first')
top_dedup.info()
```

```python
## Drop duplicates
db_dedup = db.drop_duplicates(subset = ['track_name', 'artist_name'], keep = 'first')
db_dedup.info()
```

```python
db_dedup[db_dedup.duplicated(subset = ['track_name', 'artist_name'])].sort_values(by = ['track_name'])
```

```python
## Conservative dropping
db_semi_dedup = db.drop(columns= ['genre'])
db_semi_dedup.drop_duplicates(keep = 'first', inplace = True)
db_semi_dedup.info()
```

```python
## View duplicates
db_semi_dedup[db_semi_dedup.duplicated(subset = ['track_name', 'artist_name'])].sort_values(by = ['track_name'])
```

```python
db_semi_dedup[db_semi_dedup['artist_name'] == "Howard Shore"].sort_values(by = ['track_name'])
```

Lots of songs in the 10000 per genre database were repeated in multiple genres. Additionally indexes (exa popularity) seemed to differ minorly between instances of songs. All duplicate artist + track name instance were dropped.

```python
### Get columns to match exactly
print(top.columns)
db_dedup_match = db_dedup.drop(columns = ['genre', 'mode', 'instrumentalness', 'time_signature'])
db_dedup_match['top_200'] = 0
top_dedup_match = top_dedup.drop(columns = ['highest_charting_position', 'number_of_times_charted', 'week_of_highest_charting', 'streams', 'artist_followers', 'genre_list', 'release_date', 'weeks_charted'])
```

```python
print(db_dedup_match.columns)
print(top_dedup_match.columns)
print(db_dedup_match.shape)
print(top_dedup_match.shape)
```

```python
intersection = db_dedup.merge(top, on = ['artist_name', 'track_name'], how = 'inner')
intersection.info()
```

```python
merged_dups = pd.concat([db_dedup_match, top_dedup_match], axis = 0)
merged_dups.info()
```

```python
merged_dedup = merged_dups.drop_duplicates(subset=['artist_name', 'track_name'], keep = "last")
merged_dedup.info()
```

```python
merged_dedup.head()
```

```python
merged_dedup.key.value_counts()
```

```python
# Bin similar keys together
merged_dedup['key'] = merged_dedup['key'].replace('C#/Db', 'C#')
merged_dedup['key'] = merged_dedup['key'].replace('G#/Ab', 'G#')
merged_dedup['key'] = merged_dedup['key'].replace('A#/Bb', 'A#')
merged_dedup['key'] = merged_dedup['key'].replace('F#/Gb', 'F#')
merged_dedup['key'] = merged_dedup['key'].replace('D#/Eb', 'D#')
merged_dedup.key.value_counts()
```

```python
merged_dedup.to_csv("Resources/merged_dedup.csv", index = False)
```

Current dfs:
* db = raw data of 10000 songs per genre
* top = raw data of Top 200 charting songs for 2020-2021
    * cleaned top column names to match db
* top_dedup = removing all duplicate instances of the same track_name and artist_name (there are none)
* db_dedup = removing all duplicate instances of the same track_name and artist_name
* db_semi_dedup = ignoring genre, removing all duplicate rows with any different value (there are instances of the same song+artist combo with minorly different stats i.e. popularity 41 vs 42)
* intersection = 151 song+artist combos that appear in both dfs
* merged_dups = simple concat of two dataframes, the 151 in both appear twice
* merged_dedup = 151 duplicates removed, kept the second instance where they have top_200 == 1

## DATA PREPROCESSING

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
X = merged_dedup.drop(columns = ['artist_name', 'track_name', 'track_id', 'top_200'])
y = merged_dedup['top_200']
```

```python
X_dummy = pd.get_dummies(X)
X_dummy.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_dummy, y, random_state=42)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```
